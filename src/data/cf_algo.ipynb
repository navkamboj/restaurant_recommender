```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv(r'./processed/user_rating_data.csv')
data.head()
```

```python
# add overall_rating column to this data
data['overall_rating'] = data.apply(lambda row: round((row['rating'] + row['food_rating']+ row['service_rating'])/3,2), axis=1)
data.head()
```

```python
data_matrix = pd.pivot_table(data, index=['userID'], columns=['placeID'], values=['overall_rating'])
data_matrix.head()
```

```python
# normalize matrix
data_matrix= data_matrix.subtract(data_matrix.mean(axis=1),axis='rows')
data_matrix.head()
```

```python
corr_matrix = data_matrix.T.corr(method='pearson')
corr_matrix.head()
```

```python
corr_threshold:float = 0.3
picked_user:str = 'U1001'
k_similar_users = corr_matrix[corr_matrix[picked_user]>corr_threshold][picked_user]
k_similar_users.info()
```

```python
k_similar_users
```

```python
# user that have already visited the restaurants
user_visits = data_matrix[data_matrix.index == picked_user].dropna(axis=1, how='all')

#Restaurant that similar user visited. Remove restaurants that none of the similar user have visited   
similar_user_score = data_matrix[data_matrix.index.isin(k_similar_users.index)].dropna(axis=1, how='all')

# Remove visited restaurant from the list
similar_user_score.drop(user_visits.columns,axis=1, inplace=True, errors='ignore')

similar_user_score
```

```python
# A dictionary to store item scores
item_score = {}

# Loop through items
for i in similar_user_score.columns:
    # Get the ratings for restaurant i
    restaurant_rating = similar_user_score[i]
    # Create a variable to store the score
    total = 0
    # Create a variable to store the number of scores
    count = 0
    # Loop through similar users
    for u in k_similar_users.index:
        # If the restaurant has rating
        if pd.isna(restaurant_rating[u]) == False:
            # Score is the sum of user similarity score multiply by the restaurant rating
            score = k_similar_users[u] * restaurant_rating[u]
            # Add the score to the total score for the restaurant so far
            total += score
            # Add 1 to the count
            count +=1
    # Get the average score for the item
    item_score[i] = total / count

# Convert dictionary to pandas dataframe
item_score = pd.DataFrame(item_score.items(), columns=['place', 'place_score'])
    
# Sort the restaurant by score
item_score.sort_values(by='place_score',inplace=True,ascending=False)
item_score['place'] =[i[1] for i in item_score['place']] 
# Select top m restaurant
# m = number_of_top_restaurants
item_score.reset_index(inplace=True)
item_score.drop(columns='index', inplace=True)
item_score
```
